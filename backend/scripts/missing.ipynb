this file is used to find all the problems with no description and vice versa


```python
import  json
```

```python
with open("../dsa_data.json","r",encoding="utf-8") as file:
    data = json.load(file)
with open("../dsa_descriptions.json","r") as file:
    description = json.load(file)
    
    
if data and description :
    print("data loaded 👍")
```

```python
missing_descriptions = []
for topic in data:
    for problem in data[topic]:
        if problem not in description:
            missing_descriptions.append((topic,problem))
            
print("Problems with missing description")
cnt = 0
for topic,problem in missing_descriptions:
    cnt = cnt+1
    print(f"{cnt}. {topic}: {problem}")
```

```python
missing  = []

required_fields = [
    "title",
    "description",
    "difficulty",
    "constraints",
    "examples",
    "source",
    "link"
]

for filename,problem in description.items():
    missing_fields = []
    for fields in required_fields:
        if fields not in problem:
            missing_fields.append(fields)
        elif problem[fields] == "" or problem[fields] == []:
            missing_fields.append(fields)
    if missing_fields:
        missing.append((filename,missing_fields))


print(f"Problems with missing fields: {len(missing)}\n")

for filename, fields in missing:
    print(f"{filename}")
    print(f"  Missing: {', '.join(fields)}\n")
```

```python
for problem in description.values():

    if "difficulty" not in problem:
        problem["difficulty"] = "Unknown"

    if "constraints" not in problem:
        problem["constraints"] = [
            "Constraints not available."
        ]

    if "examples" not in problem:
        problem["examples"] = [
            {
                "input": "Not available",
                "output": "Not available",
                "explanation": "Example not available for this problem."
            }
        ]
```

```python
for problem in description.values():
    if problem.get("constraints") == "Not Available":
        problem["constraints"] = [
            "Constraints not available."
        ]
```
