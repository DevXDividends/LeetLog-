this file is used to find all the problems with no description and vice versa


```python
import  json
```

```python
with open("../dsa_data.json","r",encoding="utf-8") as file:
    data = json.load(file)
with open("../dsa_descriptions.json","r") as file:
    description = json.load(file)
    
    
if data and description :
    print("data loaded 👍")
```

```python
missing_descriptions = []
for topic in data:
    for problem in data[topic]:
        if problem not in description:
            missing_descriptions.append((topic,problem))
            
print("Problems with missing description")
cnt = 0
for topic,problem in missing_descriptions:
    cnt = cnt+1
    print(f"{cnt}. {topic}: {problem}")
```
