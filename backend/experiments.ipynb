```python
from github import Auth,Github
import os 
import json
import psycopg
from psycopg.rows import dict_row
from dotenv import load_dotenv
load_dotenv() # load github token
```

```python
conn = psycopg.connect(
    dbname = "leetlog",
    user= "postgres",
    password = "root",
    host = "localhost",
    row_factory=dict_row
)
```

```python
cur = conn.cursor()
```

```python
repo = g.get_repo("DevXDividends/D.S.A")
```

```python
dsa_repo = repo.get_contents("")
```

```python
dsa_repo
```

```python
data = {}
cnt = 0
for item in dsa_repo:
    if item.type == "dir":
        print(f"\n📂 Entering Folder: {item.name}")
         # Create folder key
        data[item.name] = {}
        folder_contents = repo.get_contents(item.path)
        # print(folder_contents)
        for item_file in folder_contents:
            if item_file.type == "file":
                cnt = cnt+1
                print(f" 📄 File: {item_file.name}")
                code  = item_file.decoded_content.decode("utf-8")
                data[item.name][item_file.name] = code
                preview = "\n".join(code.splitlines()[:3])
                print(f"    Code Preview:\n{preview}\n    ...")
```

```python
cnt
```

```python
with open("dsa_data.json",'w',encoding="utf-8") as file:
    json.dump(data,file,indent=4,ensure_ascii=False)
print("All data saved !")
```
