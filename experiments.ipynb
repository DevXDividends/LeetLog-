```python
from github import Auth,Github
import os 
from dotenv import load_dotenv
load_dotenv() # load github token
```

```python
token = os.getenv("Github_access_token")
if token:
    print(True)
```

```python
auth = Auth.Token(
    token=token
)
```

```python
g = Github(
    auth=auth
)
```

```python
user = g.get_user()
```

```python
repo = user.get_repos()
```

```python
for rep in repo:
    print(rep)
```

```python
dsa_repo = g.get_repo("DevXDividends/D.S.A")
```

```python
dsa_repo_contents = dsa_repo.get_contents("")
```

```python
for item in dsa_repo_contents:
    if item.type == "dir":
        print(f"Folder: {item.name} (Path: {item.path})")
    elif item.type == "file":
        print(f"File: {item.name}")
```
